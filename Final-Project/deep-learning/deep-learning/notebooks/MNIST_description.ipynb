```python
import sys
sys.path.append("../classes")

import matplotlib.pyplot as plt
import numpy as np
import MNIST_data as mnist

%matplotlib inline
```

### Original MNIST dataset

```python
MNIST = mnist.MNIST()
X, X_valid, X_test, y, y_valid, y_test = MNIST.learning_sets()
```

```python
n1 = 2
n2 = 1000
n3 = 10
plt.imshow(X[n1,:,:], cmap="Greys_r")
plt.axis("off")
plt.savefig("../../report/figs/MNIST1.png")
plt.show()

plt.imshow(X[n2,:,:], cmap="Greys_r")
plt.axis("off")
plt.savefig("../../report/figs/MNIST2.png")
plt.show()

plt.imshow(X[n3,:,:], cmap="Greys_r")
plt.axis("off")
plt.savefig("../../report/figs/MNIST3.png")
plt.show()
```

```python
y_label = np.argmax(y.reshape((-1,10)),1)
y_valid_label = np.argmax(y_valid.reshape((-1,10)),1)
y_test_label = np.argmax(y_test.reshape((-1,10)),1)
```

```python
plt.hist(y_label, normed=True)
plt.title("Relative digits appearence in training set", fontsize=14)
plt.xlabel("digit", fontsize=15)
plt.ylabel("%", fontsize=15)
plt.xlim(0,9)
plt.savefig("../../report/figs/MNIST_Desc_Train.png")
plt.show()
```

```python
plt.hist(y_valid_label, normed=True)
plt.title("Relative digits appearence in validation set", fontsize=14)
plt.xlabel("digit", fontsize=15)
plt.ylabel("%", fontsize=15)
plt.xlim(0,9)
plt.savefig("../../report/figs/MNIST_Desc_Valid.png")
plt.show()
```

```python
plt.hist(y_test_label, normed=True)
plt.title("Relative digits appearence in test set", fontsize=14)
plt.xlabel("digit", fontsize=15)
plt.ylabel("%", fontsize=15)
plt.xlim(0,9)
plt.savefig("../../report/figs/MNIST_Desc_Test.png")
plt.show()
```

### Synthetic MNIST single digit dataset

```python
MNIST = mnist.MNIST()
X, y = MNIST.synthetize_data(nb_examples=25000, output_method="one_hot", min_length=1,
                                 max_length=1, scale_std=0.01, center_std=1.0, rotate_std=1.0)
```

```python
n1 = 2
n2 = 1000
n3 = 126
plt.imshow(X[n1,:,:,0], cmap="Greys_r")
plt.axis("off")
plt.savefig("../../report/figs/MNIST1_Synthetic.png")
plt.show()

plt.imshow(X[n2,:,:,0], cmap="Greys_r")
plt.axis("off")
plt.savefig("../../report/figs/MNIST2_Synthetic.png")
plt.show()

plt.imshow(X[n3,:,:,0], cmap="Greys_r")
plt.axis("off")
plt.savefig("../../report/figs/MNIST3_Synthetic.png")
plt.show()
```

```python
plt.hist(np.argmax(y.reshape((-1,11)),1), normed=True)
plt.xlabel("digit", fontsize=15)
plt.ylabel("%", fontsize=15)
plt.xlim(0,)
plt.show()
```
