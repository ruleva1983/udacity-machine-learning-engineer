### SVHN Single digit dataset

```python
import sys
sys.path.append("../classes")

import SVHN_data
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = SVHN_data.SVHN_Single_Digit()
X, y, X_valid, y_valid, X_test, y_test  = data.datasets(format="digits")
```

```python
plt.hist(y, normed=True)
plt.title("Relative digits appearence in training set", fontsize=14)
plt.xlabel("digit", fontsize=15)
plt.ylabel("%", fontsize=15)
plt.xlim(0,9)
plt.savefig("../../report/figs/SVHN_Desc_Train.png")
plt.show()


plt.hist(y_valid, normed=True)
plt.title("Relative digits appearence in validation set", fontsize=14)
plt.xlabel("digit", fontsize=15)
plt.ylabel("%", fontsize=15)
plt.xlim(0,9)
plt.savefig("../../report/figs/SVHN_Desc_Valid.png")
plt.show()


plt.hist(y_test, normed=True)
plt.title("Relative digits appearence in test set", fontsize=14)
plt.xlabel("digit", fontsize=15)
plt.ylabel("%", fontsize=15)
plt.xlim(0,9)
plt.savefig("../../report/figs/SVHN_Desc_Test.png")
plt.show()
```

### Multi digit SVHN dataset

```python
from six.moves import cPickle as pickle
import numpy as np
import sys
sys.path.append("../classes/")

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
pickle_file = '../SVHN_case/SVHN_MultiDigit.pickle'

with open(pickle_file, 'rb') as f:
  save = pickle.load(f)
  X_train = save['X_train']
  y_train = save['y_train']
  X_valid = save['X_valid']
  y_valid = save['y_valid']
  X_test = save['X_test']
  y_test = save['y_test']
  del save  # hint to help gc free up memory
    
    
  print('Training set', X_train.shape, y_train.shape)
  print('Validation set', X_valid.shape, y_valid.shape)
  print('Test set', X_test.shape, y_test.shape)
```

```python
np.sum(y_train[:,0] == 5)
```

```python
plt.hist(y_train[:,0], normed=False, bins=5)
plt.title("Digits with different lengths in training set", fontsize=14)
plt.xlabel("digit", fontsize=15)
plt.ylabel("Count", fontsize=15)
plt.xlim(0,5)
plt.savefig("../../report/figs/SVHN_Desc_Multi_Train.png")
plt.show()
plt.show()


plt.hist(y_valid[:,0], normed=False, bins=5)
plt.title("Digits with different lengths in validation set", fontsize=14)
plt.xlabel("digit", fontsize=15)
plt.ylabel("Count", fontsize=15)
plt.xlim(0,5)
plt.savefig("../../report/figs/SVHN_Desc_Multi_Valid.png")
plt.show()
plt.show()


plt.hist(y_test[:,0], normed=False, bins=5)
plt.title("Digits with different lengths in test set", fontsize=14)
plt.xlabel("digit", fontsize=15)
plt.ylabel("Count", fontsize=15)
plt.xlim(0,5)
plt.savefig("../../report/figs/SVHN_Desc_Multi_Test.png")
plt.show()
plt.show()
```
